```python
# Complete NSE/BSE LSTM Stock Predictor with YFinance Data Collection
# Target: 70% Accuracy

import os
import warnings
import numpy as np
import pandas as pd
from pathlib import Path
from datetime import datetime, timedelta
from typing import List, Tuple
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (Input, LSTM, Dense, Dropout, Bidirectional,
                                     BatchNormalization, Conv1D, MaxPooling1D,
                                     Concatenate, GlobalAveragePooling1D, Multiply)
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.regularizers import l1_l2
from sklearn.preprocessing import RobustScaler
from sklearn.utils import class_weight
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from tqdm import tqdm
import pickle

warnings.filterwarnings('ignore')
tf.random.set_seed(42)
np.random.seed(42)

# ============= CONFIG =============
CONFIG = {
    "DATA_DIR": Path("/content/stock_data"),
    "MODEL_DIR": Path("/content/models"),
    "WINDOW_SIZE": 45,
    "TRAIN_FRAC": 0.70,
    "VAL_FRAC": 0.15,
    "TEST_FRAC": 0.15,
    "BATCH_SIZE": 64,
    "EPOCHS": 200,
    "LEARNING_RATE": 1e-3,
    "MIN_PRICE_MOVE": 0.3,
    "DROPOUT_RATE": 0.4,
    "L1_REG": 1e-5,
    "L2_REG": 1e-4,
    "YEARS_OF_DATA": 5,  # Download 5 years of historical data
}

CONFIG['DATA_DIR'].mkdir(parents=True, exist_ok=True)
CONFIG['MODEL_DIR'].mkdir(parents=True, exist_ok=True)

# ============= NSE/BSE STOCK LISTS =============

# Top NSE stocks (NIFTY 50 components)
NSE_STOCKS = [
    'RELIANCE.NS', 'TCS.NS', 'HDFCBANK.NS', 'INFY.NS', 'ICICIBANK.NS',
    'HINDUNILVR.NS', 'ITC.NS', 'SBIN.NS', 'BHARTIARTL.NS', 'KOTAKBANK.NS',
    'LT.NS', 'AXISBANK.NS', 'ASIANPAINT.NS', 'MARUTI.NS', 'HCLTECH.NS',
    'BAJFINANCE.NS', 'WIPRO.NS', 'SUNPHARMA.NS', 'ULTRACEMCO.NS', 'TITAN.NS',
    'NESTLEIND.NS', 'ONGC.NS', 'NTPC.NS', 'TECHM.NS', 'POWERGRID.NS',
    'M&M.NS', 'TATAMOTORS.NS', 'TATASTEEL.NS', 'ADANIENT.NS', 'ADANIPORTS.NS',
    'BAJAJFINSV.NS', 'COALINDIA.NS', 'DIVISLAB.NS', 'DRREDDY.NS', 'EICHERMOT.NS',
    'GRASIM.NS', 'HEROMOTOCO.NS', 'HINDALCO.NS', 'INDUSINDBK.NS', 'JSWSTEEL.NS',
    'APOLLOHOSP.NS', 'BRITANNIA.NS', 'CIPLA.NS', 'TATACONSUM.NS', 'SBILIFE.NS',
    'HDFCLIFE.NS', 'BAJAJ-AUTO.NS', 'BPCL.NS', 'UPL.NS', 'LTIM.NS'
]

# Top BSE stocks
BSE_STOCKS = [
    'RELIANCE.BO', 'TCS.BO', 'HDFCBANK.BO', 'INFY.BO', 'ICICIBANK.BO',
    'HINDUNILVR.BO', 'ITC.BO', 'SBIN.BO', 'BHARTIARTL.BO', 'KOTAKBANK.BO',
    'LT.BO', 'AXISBANK.BO', 'ASIANPAINT.BO', 'MARUTI.BO', 'HCLTECH.BO',
    'BAJFINANCE.BO', 'WIPRO.BO', 'SUNPHARMA.BO', 'ULTRACEMCO.BO', 'TITAN.BO',
]

# Combine both (you can choose one or both)
ALL_STOCKS = NSE_STOCKS + BSE_STOCKS

# ============= DATA COLLECTION FROM YFINANCE =============

def download_stock_data(ticker: str, years: int = 5) -> pd.DataFrame:
    """Download stock data from yfinance"""
    try:
        end_date = datetime.now()
        start_date = end_date - timedelta(days=365 * years)

        stock = yf.Ticker(ticker)
        df = stock.history(start=start_date, end=end_date)

        if df.empty or len(df) < 100:
            print(f"  ⚠️  {ticker}: Insufficient data")
            return None

        # Rename columns to lowercase
        df.columns = [col.lower() for col in df.columns]
        df = df.reset_index()
        df.columns = [col.lower() for col in df.columns]

        # Add ticker column
        df['ticker'] = ticker.split('.')[0]

        # Keep only required columns
        required_cols = ['date', 'open', 'high', 'low', 'close', 'volume', 'ticker']
        df = df[[col for col in required_cols if col in df.columns]]

        return df

    except Exception as e:
        print(f"  ❌ {ticker}: {str(e)}")
        return None


def collect_all_stock_data(stock_list: List[str], data_dir: Path, years: int = 5):
    """Download and save all stock data"""
    print("\n" + "="*80)
    print(f"DOWNLOADING {len(stock_list)} STOCKS FROM YFINANCE")
    print("="*80)

    successful = 0
    failed = 0

    for ticker in tqdm(stock_list, desc="Downloading stocks"):
        # Check if already downloaded
        ticker_name = ticker.split('.')[0]
        save_path = data_dir / f"{ticker_name}.csv"

        if save_path.exists():
            print(f"  ✓ {ticker}: Already exists, skipping")
            successful += 1
            continue

        df = download_stock_data(ticker, years)

        if df is not None and len(df) > 0:
            df.to_csv(save_path, index=False)
            successful += 1
            print(f"  ✓ {ticker}: {len(df)} rows saved")
        else:
            failed += 1

    print(f"\n✓ Downloaded: {successful} stocks")
    print(f"✗ Failed: {failed} stocks")
    print(f"✓ Data saved to: {data_dir}")

    return successful, failed


# ============= FEATURE ENGINEERING =============

def add_technical_indicators(df: pd.DataFrame) -> pd.DataFrame:
    """Add comprehensive technical indicators"""
    df = df.copy()

    # Basic returns
    df['returns'] = df['close'].pct_change()
    df['log_returns'] = np.log(df['close'] / df['close'].shift(1))
    df['high_low_range'] = (df['high'] - df['low']) / df['close']
    df['close_open_ratio'] = (df['close'] - df['open']) / (df['open'] + 1e-10)

    # Volume features
    df['volume_ma5'] = df['volume'].rolling(5).mean()
    df['volume_ma20'] = df['volume'].rolling(20).mean()
    df['volume_ratio'] = df['volume'] / (df['volume_ma20'] + 1)
    df['volume_change'] = df['volume'].pct_change()
    df['price_volume'] = df['close'] * df['volume']

    # Moving averages
    for period in [5, 10, 20, 50, 100]:
        df[f'sma_{period}'] = df['close'].rolling(period).mean()
        df[f'ema_{period}'] = df['close'].ewm(span=period, adjust=False).mean()
        df[f'price_to_sma_{period}'] = (df['close'] - df[f'sma_{period}']) / (df[f'sma_{period}'] + 1e-10)

    # Crossovers
    df['sma_cross_5_20'] = np.where(df['sma_5'] > df['sma_20'], 1, -1)
    df['sma_cross_10_50'] = np.where(df['sma_10'] > df['sma_50'], 1, -1)

    # Volatility
    for window in [5, 10, 20]:
        df[f'volatility_{window}'] = df['returns'].rolling(window).std()

    # Bollinger Bands
    sma_20 = df['close'].rolling(20).mean()
    std_20 = df['close'].rolling(20).std()
    df['bb_upper_20'] = sma_20 + (2 * std_20)
    df['bb_lower_20'] = sma_20 - (2 * std_20)
    df['bb_position_20'] = (df['close'] - df['bb_lower_20']) / (df['bb_upper_20'] - df['bb_lower_20'] + 1e-10)
    df['bb_width_20'] = (df['bb_upper_20'] - df['bb_lower_20']) / (sma_20 + 1e-10)

    # RSI
    delta = df['close'].diff()
    gain = delta.where(delta > 0, 0).rolling(14).mean()
    loss = -delta.where(delta < 0, 0).rolling(14).mean()
    rs = gain / (loss + 1e-10)
    df['rsi_14'] = 100 - (100 / (1 + rs))
    df['rsi_oversold'] = (df['rsi_14'] < 30).astype(int)
    df['rsi_overbought'] = (df['rsi_14'] > 70).astype(int)

    # MACD
    ema_12 = df['close'].ewm(span=12, adjust=False).mean()
    ema_26 = df['close'].ewm(span=26, adjust=False).mean()
    df['macd'] = ema_12 - ema_26
    df['macd_signal'] = df['macd'].ewm(span=9, adjust=False).mean()
    df['macd_histogram'] = df['macd'] - df['macd_signal']

    # Momentum
    for period in [5, 10, 20]:
        df[f'momentum_{period}'] = df['close'] - df['close'].shift(period)
        df[f'roc_{period}'] = ((df['close'] - df['close'].shift(period)) / (df['close'].shift(period) + 1e-10)) * 100

    # ATR
    high_low = df['high'] - df['low']
    high_close = np.abs(df['high'] - df['close'].shift())
    low_close = np.abs(df['low'] - df['close'].shift())
    ranges = pd.concat([high_low, high_close, low_close], axis=1)
    true_range = ranges.max(axis=1)
    df['atr_14'] = true_range.rolling(14).mean()
    df['atr_ratio'] = df['atr_14'] / (df['close'] + 1e-10)

    # Stochastic
    low_14 = df['low'].rolling(14).min()
    high_14 = df['high'].rolling(14).max()
    df['stoch_k'] = 100 * (df['close'] - low_14) / (high_14 - low_14 + 1e-10)
    df['stoch_d'] = df['stoch_k'].rolling(3).mean()

    # Price patterns
    df['higher_high'] = (df['high'] > df['high'].shift(1)).astype(int)
    df['lower_low'] = (df['low'] < df['low'].shift(1)).astype(int)

    # Lagged features
    for lag in [1, 2, 3, 5]:
        df[f'returns_lag_{lag}'] = df['returns'].shift(lag)
        df[f'volume_lag_{lag}'] = df['volume'].shift(lag)

    # Rolling statistics
    for window in [5, 10]:
        df[f'returns_mean_{window}'] = df['returns'].rolling(window).mean()
        df[f'returns_std_{window}'] = df['returns'].rolling(window).std()

    # Target
    df['NextDayClose'] = df['close'].shift(-1)
    df['pct_move_next'] = (df['NextDayClose'] - df['close']) / df['close'] * 100
    df['Target'] = (df['pct_move_next'] > CONFIG['MIN_PRICE_MOVE']).astype(int)

    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    return df


def shift_features(df: pd.DataFrame, exclude_cols: List[str]) -> pd.DataFrame:
    """Shift features to prevent leakage"""
    df = df.copy()
    feature_cols = [c for c in df.columns if c not in exclude_cols]
    df[feature_cols] = df[feature_cols].shift(1)
    return df


# ============= CNN-LSTM HYBRID MODEL =============

def build_hybrid_model(input_shape):
    """Hybrid CNN-LSTM model with attention"""
    inputs = Input(shape=input_shape)

    # CNN branch
    conv1 = Conv1D(64, kernel_size=3, activation='relu', padding='same',
                   kernel_regularizer=l1_l2(CONFIG['L1_REG'], CONFIG['L2_REG']))(inputs)
    conv1 = BatchNormalization()(conv1)
    conv1 = MaxPooling1D(pool_size=2)(conv1)
    conv1 = Dropout(CONFIG['DROPOUT_RATE'])(conv1)

    conv2 = Conv1D(32, kernel_size=3, activation='relu', padding='same',
                   kernel_regularizer=l1_l2(CONFIG['L1_REG'], CONFIG['L2_REG']))(conv1)
    conv2 = BatchNormalization()(conv2)
    conv2 = Dropout(CONFIG['DROPOUT_RATE'])(conv2)

    # LSTM branch
    lstm1 = Bidirectional(LSTM(64, return_sequences=True,
                               kernel_regularizer=l1_l2(CONFIG['L1_REG'], CONFIG['L2_REG'])))(inputs)
    lstm1 = BatchNormalization()(lstm1)
    lstm1 = Dropout(CONFIG['DROPOUT_RATE'])(lstm1)

    lstm2 = Bidirectional(LSTM(32, return_sequences=True,
                               kernel_regularizer=l1_l2(CONFIG['L1_REG'], CONFIG['L2_REG'])))(lstm1)
    lstm2 = BatchNormalization()(lstm2)
    lstm2 = Dropout(CONFIG['DROPOUT_RATE'])(lstm2)

    # Attention mechanism
    from tensorflow.keras.layers import Activation, Lambda
    attention_scores = Dense(1, activation='tanh')(lstm2)
    attention_weights = Activation('softmax')(attention_scores)
    attention_output = Multiply()([lstm2, attention_weights])
    attention_output = GlobalAveragePooling1D()(attention_output)

    # Combine
    cnn_flat = GlobalAveragePooling1D()(conv2)
    combined = Concatenate()([cnn_flat, attention_output])

    # Dense layers
    x = Dense(64, activation='relu', kernel_regularizer=l1_l2(CONFIG['L1_REG'], CONFIG['L2_REG']))(combined)
    x = BatchNormalization()(x)
    x = Dropout(CONFIG['DROPOUT_RATE'])(x)

    x = Dense(32, activation='relu', kernel_regularizer=l1_l2(CONFIG['L1_REG'], CONFIG['L2_REG']))(x)
    x = Dropout(CONFIG['DROPOUT_RATE'] * 0.5)(x)

    outputs = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=inputs, outputs=outputs)

    opt = tf.keras.optimizers.Adam(learning_rate=CONFIG['LEARNING_RATE'], clipnorm=1.0)
    model.compile(
        optimizer=opt,
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.AUC(name='auc'),
                tf.keras.metrics.Precision(name='precision'),
                tf.keras.metrics.Recall(name='recall')]
    )

    return model


# ============= DATA PREPARATION =============

def prepare_sequences(csv_paths, scaler, feature_cols, window_size):
    """Prepare sequences from CSV files"""
    X_list, y_list = [], []

    for path in tqdm(csv_paths, desc="Preparing sequences"):
        df = pd.read_csv(path, parse_dates=['date'])
        df = df.sort_values('date').reset_index(drop=True)

        if len(df) < window_size + 50:
            continue

        for col in feature_cols:
            if col not in df.columns:
                df[col] = 0.0

        df[feature_cols] = df[feature_cols].fillna(method='ffill').fillna(method='bfill').fillna(0)

        # Clip outliers
        for col in feature_cols:
            mean, std = df[col].mean(), df[col].std()
            if std > 0:
                df[col] = df[col].clip(mean - 4*std, mean + 4*std)

        scaled = scaler.transform(df[feature_cols].values)
        targets = df['Target'].values

        for i in range(window_size, len(scaled) - 1):
            if not np.isnan(targets[i]) and not np.any(np.isnan(scaled[i-window_size:i])):
                X_list.append(scaled[i-window_size:i])
                y_list.append(targets[i])

    return np.array(X_list, dtype=np.float32), np.array(y_list, dtype=np.int32)


# ============= MAIN PIPELINE =============

def run_complete_pipeline(download_data=True):
    """Complete pipeline: download data, train model"""
    print("="*80)
    print("NSE/BSE STOCK PREDICTOR - TARGET: 70% ACCURACY")
    print("="*80)

    # Step 1: Download data
    if download_data:
        successful, failed = collect_all_stock_data(ALL_STOCKS, CONFIG['DATA_DIR'], CONFIG['YEARS_OF_DATA'])
        if successful == 0:
            print("❌ No data downloaded. Check your internet connection.")
            return

    # Step 2: Load CSV files
    csv_files = sorted(list(CONFIG['DATA_DIR'].glob('*.csv')))
    if not csv_files:
        print(f"❌ No CSV files found in {CONFIG['DATA_DIR']}")
        return

    print(f"\n✓ Found {len(csv_files)} stock files")

    # Step 3: Feature engineering
    print("\n✓ Adding technical indicators...")
    sample_df = pd.read_csv(csv_files[0], parse_dates=['date'])
    sample_df = add_technical_indicators(sample_df)

    exclude = ['date', 'ticker', 'NextDayClose', 'pct_move_next', 'Target']
    all_features = [c for c in sample_df.columns if c not in exclude]

    # Select best features
    numeric_df = sample_df.select_dtypes(include=[np.number])
    correlations = numeric_df.corr()['Target'].abs().sort_values(ascending=False)
    feature_cols = [f for f in correlations.index if f not in exclude][:70]

    print(f"✓ Using {len(feature_cols)} best features")

    # Step 4: Process all files
    print("\n✓ Processing all stock files...")
    for path in tqdm(csv_files, desc="Adding indicators"):
        df = pd.read_csv(path, parse_dates=['date'])
        df = add_technical_indicators(df)
        df = shift_features(df, exclude)
        df = df.dropna(subset=['close', 'Target'])
        df.to_csv(path, index=False)

    # Step 5: Fit scaler
    print("\n✓ Fitting scaler...")
    all_data = []
    for path in csv_files:
        df = pd.read_csv(path)
        valid_cols = [c for c in feature_cols if c in df.columns]
        data = df[valid_cols].dropna()
        if len(data):
            all_data.append(data.values)

    scaler = RobustScaler()
    scaler.fit(np.vstack(all_data))

    # Save scaler
    with open(CONFIG['MODEL_DIR'] / 'scaler.pkl', 'wb') as f:
        pickle.dump(scaler, f)

    # Step 6: Create sequences
    print("\n✓ Creating sequences...")
    X_all, y_all = prepare_sequences(csv_files, scaler, feature_cols, CONFIG['WINDOW_SIZE'])

    if len(X_all) == 0:
        print("❌ No sequences created")
        return

    print(f"\n✓ Total sequences: {len(X_all)}")
    print(f"  UP: {np.sum(y_all)} ({100*np.mean(y_all):.1f}%)")
    print(f"  DOWN: {len(y_all)-np.sum(y_all)} ({100*(1-np.mean(y_all)):.1f}%)")

    # Step 7: Split data
    n = len(X_all)
    train_end = int(n * CONFIG['TRAIN_FRAC'])
    val_end = int(n * (CONFIG['TRAIN_FRAC'] + CONFIG['VAL_FRAC']))

    X_train, y_train = X_all[:train_end], y_all[:train_end]
    X_val, y_val = X_all[train_end:val_end], y_all[train_end:val_end]
    X_test, y_test = X_all[val_end:], y_all[val_end:]

    print(f"\n✓ Split: Train={len(X_train)}, Val={len(X_val)}, Test={len(X_test)}")

    # Step 8: Class weights
    cw = class_weight.compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
    cw = np.clip(cw, 0.5, 2.0)
    class_weights = {i: cw[i] for i in range(len(cw))}
    print(f"✓ Class weights: {class_weights}")

    # Step 9: Build and train
    print("\n" + "="*80)
    print("TRAINING MODEL")
    print("="*80)

    model = build_hybrid_model((CONFIG['WINDOW_SIZE'], X_train.shape[2]))
    model.summary()

    callbacks = [
        EarlyStopping(monitor='val_accuracy', patience=30, restore_best_weights=True, mode='max', verbose=1),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=10, min_lr=1e-7, verbose=1),
        ModelCheckpoint(str(CONFIG['MODEL_DIR'] / 'best_model.keras'), monitor='val_accuracy',
                       save_best_only=True, mode='max')
    ]

    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=CONFIG['EPOCHS'],
        batch_size=CONFIG['BATCH_SIZE'],
        class_weight=class_weights,
        callbacks=callbacks,
        verbose=1
    )

    # Step 10: Evaluate
    print("\n" + "="*80)
    print("TEST SET EVALUATION")
    print("="*80)

    test_metrics = model.evaluate(X_test, y_test, verbose=0)
    print(f"\n✓ Test Accuracy: {test_metrics[1]*100:.2f}%")
    print(f"✓ Test AUC: {test_metrics[2]:.4f}")
    print(f"✓ Test Precision: {test_metrics[3]:.4f}")
    print(f"✓ Test Recall: {test_metrics[4]:.4f}")

    y_pred_probs = model.predict(X_test, verbose=0).flatten()
    y_pred = (y_pred_probs > 0.5).astype(int)

    print("\n" + classification_report(y_test, y_pred, target_names=['DOWN', 'UP']))

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['DOWN', 'UP'], yticklabels=['DOWN', 'UP'])
    plt.title(f'Confusion Matrix (Acc: {test_metrics[1]*100:.1f}%)')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.savefig(CONFIG['MODEL_DIR'] / 'confusion_matrix.png', dpi=200, bbox_inches='tight')
    plt.close()

    # Training plots
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))

    axes[0,0].plot(history.history['accuracy'], label='Train')
    axes[0,0].plot(history.history['val_accuracy'], label='Val')
    axes[0,0].set_title('Accuracy')
    axes[0,0].legend()
    axes[0,0].grid(True)

    axes[0,1].plot(history.history['loss'], label='Train')
    axes[0,1].plot(history.history['val_loss'], label='Val')
    axes[0,1].set_title('Loss')
    axes[0,1].legend()
    axes[0,1].grid(True)

    axes[1,0].plot(history.history['auc'], label='Train')
    axes[1,0].plot(history.history['val_auc'], label='Val')
    axes[1,0].set_title('AUC')
    axes[1,0].legend()
    axes[1,0].grid(True)

    axes[1,1].plot(history.history['precision'], label='Precision')
    axes[1,1].plot(history.history['recall'], label='Recall')
    axes[1,1].set_title('Precision & Recall')
    axes[1,1].legend()
    axes[1,1].grid(True)

    plt.tight_layout()
    plt.savefig(CONFIG['MODEL_DIR'] / 'training_history.png', dpi=200, bbox_inches='tight')

    # Save
    np.save(CONFIG['MODEL_DIR'] / 'history.npy', history.history)
    model.save(CONFIG['MODEL_DIR'] / 'final_model.keras')

    with open(CONFIG['MODEL_DIR'] / 'features.txt', 'w') as f:
        for feat in feature_cols:
            f.write(f"{feat}\n")

    print(f"\n✓ Complete! Models saved to {CONFIG['MODEL_DIR']}")
    print("="*80)

    return model, history


# ============= RUN =============
if __name__ == '__main__':
    # Set download_data=True for first run, False to skip download if data exists
    model, history = run_complete_pipeline(download_data=True)
    print("\n✓ Training complete! Check /content/models for outputs.")
```
